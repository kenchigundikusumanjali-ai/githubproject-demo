```python
lines = server_logs.split("\n")

critical_issues = []

for line in lines:
    if "[ERROR]" in line or "[CRITICAL]" in line:
        parts = line.split(" - ")
        ip = parts[0].split()[-1]
        message = parts[1]

        critical_issues.append((ip, message))

print(critical_issues)
```
